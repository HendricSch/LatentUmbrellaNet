```python
import torch
import torch.nn as nn
```

```python
class PatchGanDiscriminator(nn.Module):

    def __init__(self, in_channel: int):
        super(PatchGanDiscriminator, self).__init__()

        self.c64 = nn.Sequential(
            nn.Conv2d(in_channel, 64, kernel_size=(4, 4), stride=(2,2), padding=1),
            nn.LeakyReLU(0.2)
        )

        self.c128 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=(4, 4), stride=(2,2), padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2)
        )

        self.c256 = nn.Sequential(
            nn.Conv2d(128, 256, kernel_size=(4, 4), stride=(2,2), padding=1),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2)
        )

        self.c512 = nn.Sequential(
            nn.Conv2d(256, 512, kernel_size=(4, 4), stride=(2,2), padding=1),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2)
        )

        self.second_last = nn.Sequential(
            nn.Conv2d(512, 512, kernel_size=(4, 4), stride=(1,1), padding="same"),
            nn.BatchNorm2d(512),
            nn.LeakyReLU(0.2)
        )

        self.output = nn.Sequential(
            nn.Conv2d(512, 1, kernel_size=(4, 4), stride=(1,1), padding="same"),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.c64(x)
        x = self.c128(x)
        x = self.c256(x)
        x = self.c512(x)
        x = self.second_last(x)
        x = self.output(x)

        return x

discriminator = PatchGanDiscriminator(5)
sample = torch.randn(1, 5, 128, 64)
output = discriminator(sample)
print(output.shape)
```
